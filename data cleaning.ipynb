```python
import pandas as pd

# Define file paths
original_file_path = r"D:\Steven Documents\Project\HK options stock analysis\Stock data\0005.xlsx"
cleaned_file_path = r"D:\Steven Documents\Project\HK options stock analysis\Stock data\0005_cleaned.xlsx"

# Load the Excel file into a DataFrame
df = pd.read_excel(original_file_path)

# Remove rows with any missing values (NaN)
df = df.dropna()

# Remove rows where any column contains a zero
df = df.loc[~(df == 0).any(axis=1)]

# Print first five rows for verification
print(df.head())

# Save cleaned data as a new file instead of overwriting the original file
df.to_excel(cleaned_file_path, index=False, sheet_name="Cleaned Data")

print("Data cleaning complete. New file saved as:", cleaned_file_path)
```

```python
import os
import pandas as pd

# Define directory where stock data files are stored
stock_data_folder = r"D:\Steven Documents\Project\HK options stock analysis\Stock data"

# List all Excel files in the directory
stock_files = [f for f in os.listdir(stock_data_folder) if f.endswith(".xlsx")]

# Function to clean stock data
def clean_stock_data(file_name):
    file_path = os.path.join(stock_data_folder, file_name)
    
    # Load Excel file
    df = pd.read_excel(file_path)

    # Remove rows with any missing values (NaN)
    df_cleaned = df.dropna()

    # Remove rows where any column contains a zero
    df_cleaned = df_cleaned.loc[~(df_cleaned == 0).any(axis=1)]
    
    # Define cleaned file path (appending "_cleaned" to the filename)
    cleaned_file_name = file_name.replace(".xlsx", "_cleaned.xlsx")
    cleaned_file_path = os.path.join(stock_data_folder, cleaned_file_name)
    
    # Save cleaned data to a new Excel file
    df_cleaned.to_excel(cleaned_file_path, index=False, sheet_name="Cleaned Data")
    
    print(f"Cleaned file saved: {cleaned_file_name}")

# Loop through all stock files and clean them
for stock_file in stock_files:
    clean_stock_data(stock_file)

print("All stock files have been cleaned and saved!")
```
